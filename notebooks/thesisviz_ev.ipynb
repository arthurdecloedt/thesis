```
import os
p_n = [(f.path, f.name) for f in os.scandir('./res') if f.is_file()]
names, paths = [],[]

for p in p_n:
  names.append(p[1])
  paths.append(p[0])
res = []
for a in paths:
  with open(a, 'rb') as file:
      res.append(pickle.load(file))

names_c = [ " ".join(n.split('.')[0].split("_")[2:]) for n in names ]
print( names_c )
print(res)
keys = list(res[0][0].keys())
keys_m = list(res[0][-1].keys())
print(keys,keys_m)
res_l =[]
res_m = []
for r , name in zip(res,names_c):
  res_l.append([])
  for fold in r[:-1]:
    res_l[-1].append([fold[key] for key in keys])
  res_m.append([r[-1][key] for key in keys_m])
res_l_np = np.array(res_l)
res_m_np = np.array(res_m)
res_np.shape
```
